```python
#cargar las librerias
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#cargar los datos
datos=pd.read_excel('dataset_calificaciones.xlsx')
datos
```

```python
datos.describe()
```

```python
a = datos["A"]
b = datos["B"]
ri = a.quantile(.75)-a.quantile(.25)
print("Rango intercuantil del grupo A=", ri)
print("Límite superior del grupo A=", a.quantile(.75)+1.5*ri)
print("Límite inferior del grupo A=", a.quantile(.25)-1.5*ri)
print("----------------------------------------------------------------------")

ri2 = b.quantile(.75)-b.quantile(.25)
print("Rango intercuantil del grupo B=", ri2)
print("Límite superior del grupo B=", b.quantile(.75)+1.5*ri2)
print("Límite inferior del grupo B=", b.quantile(.25)-1.5*ri2)

datos.boxplot()
plt.show()
```

```python
conjunto_b_filtrado= b[b>1.6977763306555085]
conjunto_b_filtrado.plot(kind="box")
plt.show()
```
