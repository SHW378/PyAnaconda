```python
#cargar librerias
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#cargar los datos
datos=pd.read_excel("datos_cuantitativos.xlsx")
datos.head(10)
```

```python
edad=datos["Edad"]
sueldos=datos["sueldo"]
sueldos
```

```python
#Generar los histogramas de las variables Edad y sueldo
edad.plot(kind='hist', edgecolor='black', color='lightgreen')
plt.title("Histograma de la variable de la edad")
plt.xlabel("Clases Edad")
plt.ylabel("Frecuencia")
plt.show()

sueldos.plot(kind='hist', edgecolor='black', color='lightblue')
plt.title("Histograma de la variable del sueldo")
plt.xlabel("Clases sueldo")
plt.ylabel("Frecuencia")
plt.show()
```

```python
#generar la tabla de frecuencia de la variable edad
import math as mt
k=round(1+3.3*mt.log10(len(edad)),0)
clases_edad=pd.interval_range(start=min(edad), end=max(edad), periods=k, closed='left')
frecuencia=pd.cut(edad,clases_edad)
frecuencia=frecuencia.value_counts(sort=False)
tabla_fre_edad=pd.DataFrame(columns=["clases", "frecuencia", "frecuencia relativa", "frecuencia relativa%", "frecuencia acumulada", "frecuenica acumulada relativa%"])
tabla_fre_edad["clases"]=clases_edad
tabla_fre_edad["frecuencia"]=frecuencia.values
tabla_fre_edad["frecuencia relativa"]=(1/sum(tabla_fre_edad["frecuencia"]))*tabla_fre_edad["frecuencia"]
tabla_fre_edad["frecuencia relativa%"]=100*tabla_fre_edad["frecuencia relativa"]
tabla_fre_edad["frecuencia acumulada"]=tabla_fre_edad["frecuencia"].cumsum()
tabla_fre_edad["frecuenica acumulada relativa%"]=tabla_fre_edad["frecuencia relativa%"].cumsum()
tabla_fre_edad
```

```python
#generar tabla de frecuencia de la variable sueldos
k=round(1+3.3*mt.log10(len(sueldos)),0)
clases_sueldos=pd.interval_range(start=min(sueldos), end=max(sueldos), periods=k, closed='left')
frecuencia2=pd.cut(sueldos,clases_sueldos)
frecuencia2=frecuencia2.value_counts(sort=False)
tabla_fre_sueldos=pd.DataFrame(columns=["clases", "frecuencia", "frecuencia relativa", "frecuencia relativa%", "frecuencia acumulada", "frecuencia acumulada relativa%"])
tabla_fre_sueldos["clases"]=clases_sueldos
tabla_fre_sueldos["frecuencia"]=frecuencia2.values
tabla_fre_sueldos["frecuencia relativa"]=(1/sum(tabla_fre_sueldos["frecuencia"]))*tabla_fre_sueldos["frecuencia"]
tabla_fre_sueldos["frecuencia relativa%"]=100*tabla_fre_sueldos["frecuencia relativa"]
tabla_fre_sueldos["frecuencia acumulada"]=tabla_fre_sueldos["frecuencia"].cumsum()
tabla_fre_sueldos["frecuencia acumulada relativa%"]=tabla_fre_sueldos["frecuencia relativa%"].cumsum()
tabla_fre_sueldos
```

```python
#Medidas de tendencia central para los datos de la variable edad
print("La media de los datos de a la variable edad es:", edad.mean())
print("La mediana de los datos de a la variable edad es:", edad.median())
print("La moda de los datos de a la variable edad es:", edad.mode())
```

```python
#Medidas de tendencia central para los datos de la variable sueldo
print("La media de los datos de a la variable sueldo es:", sueldos.mean())
print("La mediana de los datos de a la variable sueldo es:", sueldos.median())
print("La moda de los datos de a la variable sueldo es:", sueldos.mode())
```

```python
#Medidads de dispercion para la variable edad
print("Rango en los datos de la variable edad=", edad.max()-edad.min())
print("Desviacion muestral en los datos de la variable edad=",edad.std())
print("Desviacion estandar poblacional en los datos de la varisable edad=", edad.std(ddof=0))
print("varianza muestral en los datos de la variable edad=", edad.var())
print("Varianza poblacional en los datos de la variable edad=", edad.var(ddof=0))
print("--------------------------------------------------------------------------------------------")
#Medidads de dispercion para la variable sueldos
print("Rango en los datos de la variable sueldos=", sueldos.max()-sueldos.min())
print("Desviacion muestral en los datos de la variable sueldos=", sueldos.std())
print("Desviacion estandar poblacional en los datos de la varisable sueldos=", sueldos.std(ddof=0))
print("varianza muestral en los datos de la variable sueldos=", sueldos.var())
print("Varianza poblacional en los datos de la variable sueldos=", sueldos.var(ddof=0))
```

```python
#Rango intercuantil y boxplot de la variable edad
ri=edad.quantile(.75)-edad.quantile(.25)
print("Rando intercuantil=", ri)
print("Límite superior=", edad.quantile(.75)+1.5*ri)
print("Límite superior=", edad.quantile(.75)-1.5*ri)

edad.plot(kind='box', color='lightgreen')
plt.show
```
