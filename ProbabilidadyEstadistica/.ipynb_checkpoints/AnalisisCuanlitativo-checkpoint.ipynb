```python
#Cargar Librerias
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Carga de datos
datos=pd.read_excel("datos_cualitativos_color_auto.xlsx")
#Imprimir los primeros 10 registros
datos.head(5)
```

```python
#Imprimir los ultimos 5 registros
datos.tail(5)
```

```python
#Total de registros
datos.count()
```

```python
#Calcular frecuencia de clases
orden=['Rojo', 'Azul', 'Verde', 'Amarillo', 'Morado']
frecuencia_color=datos['color'].value_counts().reindex(orden)
frecuencia_ta=datos['tipo auto'].value_counts()
print(frecuencia_color)
print(frecuencia_ta)
```

```python
#Grafico de barras
frecuencia_color.plot(kind='bar', edgecolor='black', color='green')

plt.title('Grafico de barra para la variable de color de Auto')
plt.xlabel=('Clases')
plt.ylabel=('Frecuencia')
plt.show
```

```python
frecuencia_ta.plot(kind='bar', edgecolor='black', color='blue')

plt.title('Grafico de barra para la variable de tipo de auto')
plt.xlabel=('Clases')
plt.ylabel=('Frecuencia')
plt.show
```

```python
#grafico de pay
frecuencia_color.plot(kind='pie', autopct='%1.2f%%')
plt.title('Grafico pie para la variable color de auto')
plt.ylabel=('Frecuencia')
plt.xlabel=('Clases')
plt.show
```

```python
frecuencia_ta.plot(kind='pie', autopct='%1.2f%%')
plt.title('Grafico pie para la variable tipo de auto')
plt.ylabel=('Frecuencia')
plt.xlabel=('Clases')
plt.show
```

```python
#Código para crear la tabla de frecuencia para la variable color 
tabla_frecuencia_color=pd.DataFrame(columns=['clases','frecuencia','frecuencia relativa', 'frecuencia relativa %', 'frecuencia acumulada', 'frecuencia acumulada %'])
clases_color_aux=frecuencia_color.index
frecuencia_color_aux=frecuencia_color.values
frecuencia_relativa_color_aux=(1/sum(frecuencia_color_aux))*frecuencia_color_aux
frecuencia_relativa_por_color_aux=frecuencia_relativa_color_aux*100
frecuencia_acum_color_aux=frecuencia_color_aux.cumsum()
frecuencia_acum_por_color_aux=frecuencia_relativa_color_aux.cumsum()

tabla_frecuencia_color['clases']=clases_color_aux
tabla_frecuencia_color['frecuencia']=frecuencia_color_aux
tabla_frecuencia_color['frecuencia relativa']=frecuencia_relativa_color_aux
tabla_frecuencia_color['frecuencia relativa %']=frecuencia_relativa_por_color_aux
tabla_frecuencia_color['frecuencia acumulada']=frecuencia_acum_color_aux
tabla_frecuencia_color['frecuencia acumulada %']=frecuencia_acum_por_color_aux
tabla_frecuencia_color
```

```python
#Código para crear la tabla de frecuencia para la variable tipo de auto
tabla_frecuencia_ta=pd.DataFrame(columns=['clases','frecuencia','frecuencia relativa', 'frecuencia relativa %', 'frecuencia acumulada', 'frecuencia acumulada %'])
clases_ta_aux=frecuencia_ta.index
frecuencia_ta_aux=frecuencia_ta.values
frecuencia_relativa_ta_aux=(1/sum(frecuencia_ta_aux))*frecuencia_ta_aux
frecuencia_relativa_por_ta_aux=frecuencia_relativa_ta_aux*100
frecuencia_acum_ta_aux=frecuencia_ta_aux.cumsum()
frecuencia_acum_por_ta_aux=frecuencia_relativa_ta_aux.cumsum()

tabla_frecuencia_ta['clases']=clases_ta_aux
tabla_frecuencia_ta['frecuencia']=frecuencia_ta_aux
tabla_frecuencia_ta['frecuencia relativa']=frecuencia_relativa_ta_aux
tabla_frecuencia_ta['frecuencia relativa %']=frecuencia_relativa_por_ta_aux
tabla_frecuencia_ta['frecuencia acumulada']=frecuencia_acum_ta_aux
tabla_frecuencia_ta['frecuencia acumulada %']=frecuencia_acum_por_ta_aux
tabla_frecuencia_ta
```
