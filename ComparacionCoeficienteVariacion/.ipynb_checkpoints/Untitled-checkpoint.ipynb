```python
#Cargar librerias
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#cargar los datos
datos=pd.read_excel("acciones_practica.xlsx")
datos.tail(5)
```

```python
#Diagramas de linea
walmex=datos["WALMEX.MX"]
walmex.plot(kind='line', color='red')
plt.title("Serie de tiempo - Precio Walmex")
plt.xlabel("Tiempo - Numero de mes")
plt.ylabel("Precio")
plt.show()

ipc=datos["^MXX"]
ipc.plot(kind='line', color='green')
plt.title("Serie de tiempo - Precio Walmex")
plt.xlabel("Tiempo - Numero de mes")
plt.ylabel("Precio")
plt.show()
```

```python
#Coeficiente de variacin
```

```python
print("Coeficiente de accion de walmex)", 100*walmex.std()/walmex.mean())
print("Coeficiente de accion de ipc)", 100*ipc.std()/ipc.mean())
```
