```python
#Cargar librerias
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Carga de datos
datos=pd.read_excel("dataset_pacientes.xlsx")
```

```python
#Calcular frecuencia de clases 
frecuencia_ts=datos['Tipo de Sangre'].value_counts()
frecuencia_cls=datos['Clasificación'].value_counts()
print(frecuencia_ts)
print(frecuencia_cls)
```

```python
#Grafico de barra 
frecuencia_ts.plot(kind='bar', edgecolor='black', color='blue')

plt.tittle='Grafico de barra para la variable de tipo de sangre'
plt.xlabel='Clases'
plt.ylabel='Frecuencia'
plt.show()
```

```python
#Grafico de barra 
frecuencia_cls.plot(kind='bar', edgecolor='black', color='blue')

plt.tittle='Grafico de barra para la variable de clasificacion'
plt.xlabel='Clases'
plt.ylabel='Frecuencia'
plt.show()
```

```python
#Grafico de pay 
frecuencia_cls.plot(kind='pie', autopct='%1.2f%%')

plt.tittle='Grafico de pay para la variable de clasificacion'
plt.xlabel='Clases'
plt.ylabel='Frecuencia'
plt.show()
```

```python
#Grafico de pay 
frecuencia_ts.plot(kind='pie', autopct='%1.2f%%')

plt.tittle='Grafico de pay para la variable de tipo de sangre'
plt.xlabel='Clases'
plt.ylabel='Frecuencia'
plt.show()
```

```python
#frecuencia relativa para la variable tipo de sangre 
tabla_frecuencia_ts=pd.DataFrame(columns=['clases','frecuencia','frecuencia relativa', 'frecuencia relativa %', 'frecuencia acumulada', 'frecuencia acumulada %'])
clases_ts_aux=frecuencia_ts.index
frecuencia_ts_aux=frecuencia_ts.values
frecuencia_relativa_ts_aux=(1/sum(frecuencia_ts_aux))*frecuencia_ts_aux
frecuencia_relativa_por_ts_aux=frecuencia_relativa_ts_aux*100
frecuencia_acum_ts_aux=frecuencia_ts_aux.cumsum()
frecuencia_acum_por_ts_aux=frecuencia_relativa_ts_aux.cumsum()*100

tabla_frecuencia_ts['clases']=clases_ts_aux
tabla_frecuencia_ts['frecuencia']=frecuencia_ts_aux
tabla_frecuencia_ts['frecuencia relativa']=frecuencia_relativa_ts_aux
tabla_frecuencia_ts['frecuencia relativa %']=frecuencia_relativa_por_ts_aux
tabla_frecuencia_ts['frecuencia acumulada']=frecuencia_acum_ts_aux
tabla_frecuencia_ts['frecuencia acumulada %']=frecuencia_acum_por_ts_aux
tabla_frecuencia_ts
```

```python
#Frecuencia relativa para la variable clasificacion
tabla_frecuencia_cls=pd.DataFrame(columns=['clases','frecuencia','frecuencia relativa', 'frecuencia relativa %', 'frecuencia acumulada', 'frecuencia acumulada %'])
clases_cls_aux=frecuencia_cls.index
frecuencia_cls_aux=frecuencia_cls.values
frecuencia_relativa_cls_aux=(1/sum(frecuencia_ts_aux))*frecuencia_cls_aux
frecuencia_relativa_por_cls_aux=frecuencia_relativa_cls_aux*100
frecuencia_acum_cls_aux=frecuencia_cls_aux.cumsum()
frecuencia_acum_por_cls_aux=frecuencia_relativa_cls_aux.cumsum()*100

tabla_frecuencia_cls['clases']=clases_cls_aux
tabla_frecuencia_cls['frecuencia']=frecuencia_cls_aux
tabla_frecuencia_cls['frecuencia relativa']=frecuencia_relativa_cls_aux
tabla_frecuencia_cls['frecuencia relativa %']=frecuencia_relativa_por_cls_aux
tabla_frecuencia_cls['frecuencia acumulada']=frecuencia_acum_cls_aux
tabla_frecuencia_cls['frecuencia acumulada %']=frecuencia_acum_por_cls_aux
tabla_frecuencia_cls
```
