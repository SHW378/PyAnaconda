```python
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
def graficarDistribucionNormalAcumulada(media, desviacion, a, b):
    valores=[]
    probabilidades=[]
    for i in range(a,b+1,1):
        valores.append(i)
        probabilidades.append(norm.cdf(i,loc=media, scale=desviacion))
    return valores, probabilidades
 
def graficarDistribucionNormal(media, desviacion, a, b):
    valores=[]
    probabilidades=[]
    for i in range(a,b+1,1):
        valores.append(i)
        probabilidades.append(norm.pdf(i,loc=media,scale=desviacion))
    return valores, probabilidades
```

Ejercicio:
La estatura de una población de adultos sigue una distribución normal con media μ=170\mu = 170μ=170 cm y desviación estándar σ=10\sigma = 10σ=10 cm.
<ol>
<li>¿Cuál es la probabilidad de que una persona seleccionada al azar mida menos de 160 cm?</li>
<li>¿Cuál es la probabilidad de que mida entre 165 cm y 185 cm?</li>
<li>¿Cuál es la probabilidad de que mida más de 180 cm?</li>
<li>¿Cuál es la probabilidad de que mida más de 200 cm?</li>
<li>¿Cuál es el valor esperado?</li>
<li>¿Cuál es la varianza</li>
</ol>

```python
print("1.- p(x<=160):", norm.cdf(160,loc=170,scale=10)) 
print("2. p(165<=X<=185)", norm.cdf(185, loc=170,scale=10)-norm.cdf(165, loc=170,scale=10))
print("3.- p(x>=180)", 1-norm.cdf(180, loc=170,scale=10))
print("4.- p(x>=200)", 1-norm.cdf(200, loc=170,scale=10))
print("5.- E(X):", 170)
print("6.- VAR(X):", 100)
print("7.- STD(X):", 10)
```

Ejercicio 2: En un equipo de desarrollo de software, el tiempo que tarda un programador en resolver un bug crítico sigue una distribución normal con una media de 10 horas y una desviación estándar de 2 horas.
 
1.- ¿Cuál es la probabilidad de que un programador resuelva el bug en menos de 8 horas?<br>
2.-¿Cuál es la probabilidad de que tarde entre 9 y 12 horas?<br>
3.- Gráficar la función de densidad y la distribución acumulada<br>
4.- ¿Cuántas horas necesita el programador para resolver el bug a una probabilidad de 95%?<br>
5.- ¿Cuántas horas necesita el programador para resolver el bug con una probabilidad de 33%?<br>

```python
print("1.- p(x<=8)", norm.cdf(8,loc=10,scale=2))
print("2.- p(9<=X<=12)", norm.cdf(12,loc=10,scale=2)-norm.cdf(9,loc=10,scale=2))
print("3.-----------------------------")
valores,f=graficarDistribucionNormal(10, 2, 0, 20)
plt.plot(valores,f)
plt.title("Función de densidad de una N(10,4)")
plt.xlabel("X")
plt.ylabel("f(x)")
plt.show()
 
valores,f=graficarDistribucionNormalAcumulada(10, 2, 0, 20)
plt.plot(valores,f)
plt.title("Función acumulada de probabilidad de una N(170,100)")
plt.xlabel("X")
plt.ylabel("P(X<=xi)")
plt.show()

print("4.- a=", norm.ppf(0.95,loc=10,scale=2)) 
print("4.- comprobar(pX<=13.29)", norm.cdf(13.29,loc=10,scale=2)) 
print("5.- a=", norm.ppf(.33,loc=10,scale=2))
print("5.- comprobar(pX<=9.12)", norm.cdf(9.12,loc=10,scale=2))
```
